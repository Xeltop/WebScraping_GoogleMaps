```python
import pandas as pd
from transformers import pipeline
from textblob import TextBlob
```

```python
df = pd.read_csv("/content/google (1).csv")
```

```python
print(df.columns)
```

```python
df_comm = df[["wiI7pd"]]
```

```python
df_comm.isnull()
```

```python
df_comm = df_comm.dropna()
```

```python
df_comm.head()
```

```python
sentiment_model = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")
```

```python
def analyze_sentiment(review):
    if not isinstance(review, str):
        return None
    result = sentiment_model(review[:512])
    label = result[0]['label']
    return label

def assign_rating(label):
    if label == '1 star':
        return 1
    elif label == '2 stars':
        return 2
    elif label == '3 stars':
        return 3
    elif label == '4 stars':
        return 4
    elif label == '5 stars':
        return 5
    return None
```

```python
df_comm['sentiment_label'] = df_comm['wiI7pd'].apply(analyze_sentiment)
df_comm['rating'] = df_comm['sentiment_label'].apply(assign_rating)

print(df_comm)
```

```python
from textblob import download_corpora
import nltk
nltk.download('brown')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt_tab')
```

```python
def extract_keywords(review):
    if not isinstance(review, str):
        return None
    blob = TextBlob(review)
    keywords = blob.noun_phrases
    return keywords

df_comm['keywords'] = df_comm['wiI7pd'].apply(extract_keywords)
print(df_comm)
```

```python
#Show examples of keywords for rating of 5 and 3 to see the diff

five_star_comments = df_comm[df_comm['rating'] == 5]
three_star_comments = df_comm[df_comm['rating'] == 3]

five_star_sample = five_star_comments[['wiI7pd', 'keywords']].tail()
three_star_sample = three_star_comments[['wiI7pd', 'keywords']].tail(5)

print("Keywords for 5-star rated comments:")
print(five_star_sample)
print("\nKeywords for 3-star rated comments:")
print(three_star_sample)
```

```python
import shap

explainer = shap.Explainer(sentiment_model)

sample_reviews = df_comm['wiI7pd'].sample(n=1, random_state=12).tolist()

shap_values = explainer(sample_reviews)

for i, explanation in enumerate(shap_values):
    print(f"\nSHAP Explanation for Sample Review {i + 1}: {sample_reviews[i]}")
    shap.plots.text(explanation)
```
